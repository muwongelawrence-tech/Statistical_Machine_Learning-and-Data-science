```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#first method to load an image 
from IPython.display import Image
filename = 'traffic.jpg'
Image(filename ,width=224, height=224)
```

```python
#second method of loading an image 
from tensorflow.keras.preprocessing import image
img = image.load_img('traffic.jpg',target_size = (224,224))
plt.imshow(img)
```

```python
# third method to load an image 
import cv2
imgg = cv2.imread(filename)
imgg = cv2.resize(imgg, (224,224))
plt.imshow(imgg)
```

```python
#fourth method to load an image
from PIL import Image #pip install pillow
im = Image.open(filename)
im = im.resize((224, 224))
plt.imshow(im)
```

```python
#lets load the deep learning Model
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
#Pre-processing of the Image
from tensorflow.keras.preprocessing import image
img = image.load_img(filename , target_size = (224, 224))
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
predictions = mobile.predict(final_image)
```
