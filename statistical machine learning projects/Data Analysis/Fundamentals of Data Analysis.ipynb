```python
# import pandas library
import pandas as pd
import numpy as np
```

```python
# Because the data does not include headers, we can add an argument headers = None inside the read_csv() 
# method so that pandas will not automatically set the first row as a header.

# Read the online file by the URL provides above, and assign it to variable "df"
other_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
df = pd.read_csv(other_path, header=None)
```

```python
# Add Headers
# Take a look at our dataset. Pandas automatically set the header with an integer starting from 0.
# Thus, we have to add headers manually.
# First, we create a list "headers" that include all column names in order.
# Then, we use dataframe.columns = headers to replace the headers with the list we created.
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# Lets replace headers and recheck our dataframe:
df.columns = headers
df.head(10)
```

```python
# We need to replace the "?" symbol with NaN so the dropna() can remove the missing values:
df1=df.replace('?',np.NaN)
```

```python
df1.head(10)
```

```python
# We can drop missing values along the column "price" as follows:
df=df1.dropna(subset=["price"], axis=0)
df.head(10)
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df2 = pd.read_csv(filename, names = headers)
```

```python
import numpy as np

# replace "?" to NaN
df2.replace("?", np.nan, inplace = True)
df2.head(5)
```

```python
# Evaluating for Missing Data
# The missing values are converted by default. We use the following functions to identify these missing values.
# There are two methods to detect missing data:
# .isnull()
# .notnull()
# The output is a boolean value indicating whether the value that is passed into the argument is in fact
# missing data.
missing_data = df2.isnull()
missing_data.head(5)
```

```python
# "True" means the value is a missing value while "False" means the value is not a missing value.
# Count missing values in each column
# Using a for loop in Python, we can quickly figure out the number of missing values in each column.
# As mentioned above, "True" represents a missing value and "False" means the value is present in the dataset. 
# In the body of the for loop the method ".value_counts()" counts the number of "True" values.
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")  
```

```python
# Deal with missing data
# How to deal with missing data?
# Drop data
# a. Drop the whole row
# b. Drop the whole column
# Replace data
# a. Replace it by mean
# b. Replace it by frequency
# c. Replace it based on other functions
# Calculate the mean value for the "normalized-losses" column 
avg_norm_loss = df2["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
# Replace "NaN" with mean value in "normalized-losses" column
df2["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
df2.head()
```

```python
# Calculate the mean value for the "bore" column
avg_bore=df2['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df2["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df2["stroke"].astype('float').mean(axis=0)
print("Average of stroke: ", avg_stroke)
```

```python
df2["stroke"].replace(np.nan , avg_stroke, inplace=True)
```

```python
df2["stroke"].describe()
```

```python
avg_price = df2["price"].astype('float').mean(axis=0)
print("Average of price: ", avg_price)

df2["price"].replace(np.nan , avg_price, inplace=True)

df2["price"].describe()
```

```python
avg_horsepower = df2['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)

df2['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df2['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)

df2['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```
